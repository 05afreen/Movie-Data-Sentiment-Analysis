```python
#Oscar column cleaning
df['Oscar'] = df['Oscar'].apply(lambda x: int(re.search(r'\d+', str(x)).group()) if re.search(r'\d+', str(x)) else None)
```

```python
df['Oscar'] = df['Oscar'].replace([np.nan, np.inf, -np.inf], 0)

# Convert float to int
df['Oscar'] = df['Oscar'].astype(int)
```

```python
#nomination column cleaning 
df[['Wins', 'Nominations']] = df['Awards'].str.extract(r'(\d+)\s+wins\s+&\s+(\d+)\s+nominations')

df['Wins'] = df['Wins'].fillna(0).astype(int)
df['Nominations'] = df['Nominations'].fillna(0).astype(int)
# Convert extracted values to integers
df['Wins'] = df['Wins'].astype(int)
df['Nominations'] = df['Nominations'].astype(int)
```

```python
#metascore clumn cleaning
df["MetaScore"] = pd.to_numeric(df['MetaScore'], errors='coerce').fillna(0).astype(int)
```

```python
#rating count column cleaning
def convert_k_to_m(value):
    if isinstance(value, str):
        if 'K' in value:  # Check if the value contains 'K'
            return float(value.replace('K', '').strip()) / 1000  # Convert 'K' to number and divide by 1000
        elif 'M' in value:  # If already 'M', just return the number
            return float(value.replace('M', '').strip())
    return value  # If it's neither
```

```python
#rating_count cleaning
df['Rating_count(M)'] = df['Rating_count'].apply(convert_k_to_m)
```

```python
df.drop("Rating_count",axis=1,inplace=True)
```

```python
# duration column cleaning 
df['Duration'] = df['Duration'].astype(str)
```

```python
df['Duration'] = df['Duration'].str.replace('h', ':').str.replace('m', ' ')
```

```python
def convert_to_timedelta(duration):
    if pd.isna(duration) or not duration.strip():  # Handle NaN and empty strings
        return pd.NaT
    # Replace the space and convert to proper format
    duration = duration.replace(' ', '')  # Remove any space between hours and minutes
    try:
        hours, minutes = map(int, duration.split(':'))  # Split and convert to integers
        return pd.Timedelta(hours=hours, minutes=minutes)
    except ValueError:
        return pd.NaT  # Return NaT for any invalid format

# Apply the conversion function to the Duration column
df['Duration'] = df['Duration'].apply(convert_to_timedelta)
```

```python
 #  Top 10 Grossing Movies
top_grossing_movies = df[['Movie Name', 'Gross World Wide']].sort_values(by='Gross World Wide', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10,6))
sns.barplot(x='Gross World Wide', y='Movie Name', data=top_grossing_movies, palette='Blues_d')
plt.title("Top 10 Grossing Movies")
plt.xlabel("Gross Worldwide Revenue (in Millions)")
plt.ylabel("Movie Name")
plt.show()
```

```python
revenue_by_director = df.groupby('Director')['Gross World Wide'].sum().sort_values(ascending=False).head(10)

# Plotting the top 10 directors by revenue
plt.figure(figsize=(12, 8))
sns.barplot(x=revenue_by_director.index, y=revenue_by_director.values, palette='viridis')

# Adding title and labels
plt.title("Top 10 Directors by Total Revenue", fontsize=16)
plt.xlabel("Director", fontsize=12)
plt.ylabel("Total Revenue (in USD)", fontsize=12)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
